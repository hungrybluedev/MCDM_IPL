```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
plt.rcParams["figure.dpi"] = 1000
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = '8'
```

```python
target = 'bowlers'
```

```python
ranking_data = pd.read_csv(target + '_ranks.csv')
ranking_data
```

```python
just_ranks = ranking_data.drop('Theta', axis=1)
just_ranks
```

```python
x = just_ranks.columns

plt.xticks(rotation='vertical')
plt.yticks(ticks=range(1, len(x) + 1))

plt.xlabel('Batsmen' if target == 'batsmen' else 'Bowlers')
plt.ylabel('Rank')

for _, row in just_ranks.iterrows():
    plt.scatter(x=x, y=list(row), alpha=0.002, c='black')
```
